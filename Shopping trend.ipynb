```python
# perform chi square test to see if there is any correlation between Subsription status
crosstabresult = pd.crosstab(index = df1['Frequency of Purchases'], columns = df1['Subscription Status'])
crosstabresult
```

```python
from scipy.stats import chi2_contingency
chisqresult = chi2_contingency(crosstabresult)
```

```python
#pvalue of chi square test
print(chisqresult[1])
```

The p value is significantly larger than 0.05, hence we can conclude that there is no correlation between subsription status and frequency of purchase

```python
def rating(row):
    if row['Review Rating']>1 and row['Review Rating']<2:
        return 1
    if row['Review Rating']>2 and row['Review Rating']<3:
        return 3
    if row['Review Rating']>3 and row['Review Rating']<4:
        return 4
    if row['Review Rating']>4 and row['Review Rating']<=5:
        return 5
   
```

```python
df1['Rating']=df1.apply(lambda row:rating(row),axis=1)
```

```python
df1['Rating'].value_counts()
```

```python
df1[df1['Rating']==5]['Category'].value_counts()
```

```python
sns.boxplot(x='Category',y='Review Rating', data = df1)
```
